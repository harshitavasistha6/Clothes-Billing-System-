```python
from tkinter import *
from tkinter import messagebox
import random,os,tempfile
```

```python
# here is function

def print_bill():
    if textarea.get(1.0,END)=="\n":
        messagebox.showerror("Error","Bill is Empty")
    else:   
        file=tempfile.mktemp(".txt")
        open(file,"w").write(textarea.get(1.0,END))
        os.startfile(file,'print')
        
        



def search_bill():
    for i in os.listdir("bills/"):
        if i.split('.')[0]==billentry.get():
            f=open(f"bills/{i}","r")
            textarea.delete(1.0,END)
            for data in f:
                textarea.insert(END,data)
            f.close() 
            break
    else: 
        messagebox.showerror("Error","Invalid Bill Number")

if not os.path.exists('bills'):
    os.mkdir("bills")

def savebill():
    global billnumber
    result=messagebox.askyesno("confirm","Do you want to save the bill?")
    if result:
        bill_content=textarea.get(1.0,END)
        file =open(f'bills/ {billnumber}.txt','w')
        file.write(bill_content)
        file.close()
        messagebox.showinfo("success",f" billnumber{billnumber} is saved successfully")
        billnumber=random.randint(500,1000)
        
    
billnumber=random.randint(500,1000)


def bill():
    if nameentry.get()=="" or phoneentry.get()=="" :
        messagebox.showerror("Error","Customer Details is Requried")
        
        
        
    else:
        textarea.delete(1.0,END)  #  old deleted everthing when new added
        textarea.insert(END,"\t \t ***Welcome***\n")
        textarea.insert(END,f"Bill Number:  {billnumber}\n")
        textarea.insert(END,f"Customer Name: {nameentry.get()}\n")
        textarea.insert(END,f"Phone Number: {phoneentry.get()}\n")
        textarea.insert(END,"**********************************************************\n")
        textarea.insert(END,"Product \t \t  \t  Quntity   \t \t \t Prize \n")
        textarea.insert(END,"**********************************************************\n")
        if shirtentry.get()!='0':
            textarea.insert(END,f"Shirt\t \t \t{shirtentry.get()}\t \t \t {shirtprize} Rs\n")
        if tshirtentry.get()!='0':
            textarea.insert(END,f"TShirt\t \t \t{tshirtentry.get()}\t \t \t {tshirtprize} Rs\n") 
        if pentsentry.get()!='0':
            textarea.insert(END,f"Pents\t \t \t{pentsentry.get()}\t \t \t {pentprize} Rs\n") 
        if trousersentry.get()!='0':
            textarea.insert(END,f"Trouser\t \t \t{trousersentry.get()}\t \t \t {trouserprize} Rs\n") 
        if joggersentry.get()!='0':
            textarea.insert(END,f"Joggers\t \t \t{joggersentry.get()}\t \t \t {joggerprize} Rs\n")
        if formalentry.get()!='0':
            textarea.insert(END,f"Formals\t \t \t{formalentry.get()}\t \t \t {formalprize} Rs\n") 
        if topsentry.get()!='0':
            textarea.insert(END,f"Tops\t \t \t{topsentry.get()}\t \t \t {topprize} Rs\n")
        if sareeentry.get()!='0':
            textarea.insert(END,f"sarees\t \t \t{sareeentry.get()}\t \t \t {sareeprize} Rs\n")
        if jeansentry.get()!='0':
            textarea.insert(END,f"Jeans\t \t \t{jeansentry.get()}\t \t \t {jeansprize} Rs\n")    
        if suitentry.get()!='0':
            textarea.insert(END,f"Suits\t \t \t{suitentry.get()}\t \t \t {suitprize} Rs\n") 
        if gownentry.get()!='0':
            textarea.insert(END,f"Gown\t \t \t{gownentry.get()}\t \t \t {gownprize} Rs\n")
        if skirtentry.get()!='0':
            textarea.insert(END,f"Skirt\t \t \t{skirtentry.get()}\t \t \t {skirtprize}Rs\n")
        if bodysuitsentry.get()!="0":
            textarea.insert(END,f"BodySuit\t \t \t {bodysuitsentry.get()}\t \t \t {bodysuitprize}Rs \n")
        if beaniesentry.get()!="0":
            textarea.insert(END,f"Beanies\t \t \t {beaniesentry.get()}\t \t \t {beaniesprize}Rs \n") 
        if mittensentry.get()!="0":
            textarea.insert(END,f"Mittense\t \t \t {mittensentry.get()}\t \t \t {mittenprize}Rs \n")  
        if bootiesentry.get()!="0":
            textarea.insert(END,f"Booties\t \t \t {bootiesentry.get()}\t \t \t {bootiesprize}Rs \n") 
        if sweaterentry.get()!="0":
            textarea.insert(END,f"Sweater\t \t \t {sweaterentry.get()}\t \t \t {sweaterprize}Rs \n") 
        if rompersentry.get()!="0":
            textarea.insert(END,f"Romper\t \t \t {rompersentry.get()}\t \t \t {rompersprize}Rs \n")
        textarea.insert(END,"**********************************************************\n")
        
        # here is tax addded////////////////////////////////////////////////////////////////////////////////
        if meanswareprizetaxentry.get()!="0.0 Rs":
            textarea.insert(END,f"Means Ware Tax \t \t {meanswareprizetaxentry.get()}\n")
        if womenwareprizetaxentry.get()!="0.0 Rs":
            textarea.insert(END,f"Womens Ware Tax \t \t {womenwareprizetaxentry.get()}\n")  
        if childwareprizetaxentry.get()!="0.0 Rs":
            textarea.insert(END,f"Child Ware Tax \t \t {childwareprizetaxentry.get()}\n") 
         
        textarea.insert(END,f"Total Button\t \t \t \t {totalbill}")
        
        savebill()
    
def total():
    global shirtprize,tshirtprize,pentprize,trouserprize,joggerprize,formalprize,topprize,sareeprize,jeansprize,suitprize,gownprize,skirtprize,bodysuitprize,beaniesprize,mittenprize,bootiesprize,sweaterprize,rompersprize,totalbill
    shirtprize=int(shirtentry.get())*600
    tshirtprize=int(tshirtentry.get())*1200
    pentprize=int(pentsentry.get())*1000
    trouserprize=int(trousersentry.get())*1200
    joggerprize=int(joggersentry.get())*900
    formalprize=int(formalentry.get())*2000
    
    totalmeanswareprize = shirtprize+tshirtprize+pentprize+trouserprize+joggerprize+formalprize
    meanswareprizeentry.delete(0,END)
    meanswareprizeentry.insert(0,str(totalmeanswareprize)+" Rs")
    
    menswaretax=totalmeanswareprize*0.18
    meanswareprizetaxentry.delete(0,END)
    meanswareprizetaxentry.insert(0,str(menswaretax)+" Rs")
    
    
     #here is women ware............................................................................
    topprize=int(topsentry.get())*500 
    sareeprize=int(sareeentry.get())*2500
    jeansprize =int(jeansentry.get())*1200
    suitprize=int(suitentry.get())*700
    gownprize=int(gownentry.get())*2300
    skirtprize=int(skirtentry.get())*400
    
    totalwomenwareprize=topprize+sareeprize+jeansprize+suitprize+gownprize+skirtprize
    womenswareprizeentry.delete(0,END)
    womenswareprizeentry.insert(0,str(totalwomenwareprize)+" Rs")
    
    womenwaretax=totalwomenwareprize*0.15
    womenwareprizetaxentry.delete(0,END)
    womenwareprizetaxentry.insert(0,str(womenwaretax)+" Rs")
    
    
    
    # here is childware ....................................................................................
    bodysuitprize=int(bodysuitsentry.get())*1200
    beaniesprize=int(beaniesentry.get())*600
    mittenprize=int(mittensentry.get())*760
    bootiesprize=int(bootiesentry.get())*500
    sweaterprize=int(sweaterentry.get())*1200
    rompersprize=int(rompersentry.get())*650
    
    totalchildwareprize=bodysuitprize+beaniesprize+mittenprize+bootiesprize+sweaterprize+rompersprize
    childwareprizeentry.delete(0,END)
    childwareprizeentry.insert(0,str(totalchildwareprize)+" Rs")
    
    childwaretax=totalchildwareprize*0.10
    childwareprizetaxentry.delete(0,END)
    childwareprizetaxentry.insert(0,str(childwaretax)+ " Rs")
    
    totalbill=totalmeanswareprize+totalwomenwareprize+totalchildwareprize+menswaretax+womenwaretax+childwaretax
    
    
    
    




root=Tk()
root.title("Clothing Billing System")
root.geometry("1280x687")# widthxheight
hadinglabel=Label(root,text="Colthing Billing System",font=("Time New Roman",30,"bold"),bg="gray20",fg="gold",bd=12,relief=GROOVE)
hadinglabel.pack(fill=X)

# this is frame .......................................................................
customer_details_frame=LabelFrame(root,text="Customer Details",font=("Time New Roman",15,"bold"),bg="gray20",fg="gold",bd=8,relief=GROOVE)
customer_details_frame.pack(fill=X)

namelabel=Label(customer_details_frame,text="Name",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
namelabel.grid(row=0,column=0,padx=20)
nameentry=Entry(customer_details_frame,font=("arial",15),bd=8,width=18)
nameentry.grid(row=0,column=1,padx=8)

phonelabel=Label(customer_details_frame,text="PhoneNumber",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
phonelabel.grid(row=0,column=2,padx=20)
phoneentry=Entry(customer_details_frame,font=("arial",15),bd=8,width=18)
phoneentry.grid(row=0,column=3,padx=8)


billlabel=Label(customer_details_frame,text="BillNumber",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
billlabel.grid(row=0,column=4,padx=20)
billentry=Entry(customer_details_frame,font=("arial",15),bd=8,width=18)
billentry.grid(row=0,column=5,padx=8)

searchbutton=Button(customer_details_frame,text="Search",font=("arial",12,"bold"),bd=12,width=10,command=search_bill)
searchbutton.grid(row=0,column=6,padx=20,pady=8)
# here first part is completed///////////////////////////////////////////////////////////////////////////////////////////

# here product frame is created.........................................................................
productframe=Frame(root)
productframe.pack()

menswereframe=LabelFrame(productframe,text="MeansWare",font=("Time New Roman",15,"bold"),bg="gray20",fg="gold",bd=8,relief=GROOVE)
menswereframe.grid(row=0,column=0)

shirtlabel=Label(menswereframe,text="Shirt",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
shirtlabel.grid(row=0,column=0,pady=9)
shirtentry=Entry(menswereframe,font=("arial",15),bd=5,width=12)
shirtentry.grid(row=0,column=1,pady=9)
shirtentry.insert(0,0)


tshirtlabel=Label(menswereframe,text="Tshirt",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
tshirtlabel.grid(row=1,column=0,pady=9)
tshirtentry=Entry(menswereframe,font=("arial",15),bd=5,width=12)
tshirtentry.grid(row=1,column=1,pady=9)
tshirtentry.insert(0,0)

pentslabel=Label(menswereframe,text="Pents",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
pentslabel.grid(row=2,column=0,pady=9)
pentsentry=Entry(menswereframe,font=("arial",15),bd=5,width=12)
pentsentry.grid(row=2,column=1,pady=9)
pentsentry.insert(0,0)

trouserslabel=Label(menswereframe,text="Trousers",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
trouserslabel.grid(row=3,column=0,pady=9)
trousersentry=Entry(menswereframe,font=("arial",15),bd=5,width=12)
trousersentry.grid(row=3,column=1,pady=9)
trousersentry.insert(0,0)

joggerslabel=Label(menswereframe,text="Joggers",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
joggerslabel.grid(row=4,column=0,pady=9)
joggersentry=Entry(menswereframe,font=("arial",15),bd=5,width=12)
joggersentry.grid(row=4,column=1,pady=9)
joggersentry.insert(0,0)

formallabel=Label(menswereframe,text="Formals",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
formallabel.grid(row=5,column=0,pady=9)
formalentry=Entry(menswereframe,font=("arial",15),bd=5,width=12)
formalentry.grid(row=5,column=1,pady=9)
formalentry.insert(0,0)

# here is women ware
womenswereframe=LabelFrame(productframe,text="WomeansWare",font=("Time New Roman",15,"bold"),bg="gray20",fg="gold",bd=8,relief=GROOVE)
womenswereframe.grid(row=0,column=2)

topslabel=Label(womenswereframe,text="Tops",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
topslabel.grid(row=0,column=2,pady=9,padx=10)
topsentry=Entry(womenswereframe,font=("arial",15),bd=5,width=12)
topsentry.grid(row=0,column=3,pady=9,padx=10)
topsentry.insert(0,0)

sareelabel=Label(womenswereframe,text="Saree",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
sareelabel.grid(row=1,column=2,pady=9,padx=10)
sareeentry=Entry(womenswereframe,font=("arial",15),bd=5,width=12)
sareeentry.grid(row=1,column=3,pady=9,padx=10)
sareeentry.insert(0,0)

jeanslabel=Label(womenswereframe,text="Jeans",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
jeanslabel.grid(row=2,column=2,pady=9)
jeansentry=Entry(womenswereframe,font=("arial",15),bd=5,width=12)
jeansentry.grid(row=2,column=3,pady=9,padx=10)
jeansentry.insert(0,0)


suitlabel=Label(womenswereframe,text="Suit",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
suitlabel.grid(row=3,column=2,pady=9,padx=10)
suitentry=Entry(womenswereframe,font=("arial",15),bd=5,width=12)
suitentry.grid(row=3,column=3,pady=9,padx=10)
suitentry.insert(0,0)


gownlabel=Label(womenswereframe,text="Gown",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
gownlabel.grid(row=4,column=2,pady=9,padx=10)
gownentry=Entry(womenswereframe,font=("arial",15),bd=5,width=12)
gownentry.grid(row=4,column=3,pady=9,padx=10 )
gownentry.insert(0,0)

skirtlabel=Label(womenswereframe,text="Skirt",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
skirtlabel.grid(row=5,column=2,pady=9)
skirtentry=Entry(womenswereframe,font=("arial",15),bd=5,width=12)
skirtentry.grid(row=5,column=3,pady=9,padx=10)
skirtentry.insert(0,0)

# here is child section

childwareframe=LabelFrame(productframe,text="ChildWareFrame",font=("Time New Roman",15,"bold"),bg="gray20",fg="gold",bd=8,relief=GROOVE)
childwareframe.grid(row=0,column=4)

bodysuitslabel=Label(childwareframe,text="BodySuits",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
bodysuitslabel.grid(row=0,column=4,pady=9)
bodysuitsentry=Entry(childwareframe,font=("arial",15),bd=5,width=12)
bodysuitsentry.grid(row=0,column=5,pady=9,padx=10)
bodysuitsentry.insert(0,0)



beanielabel=Label(childwareframe,text="Beanie",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
beanielabel.grid(row=1,column=4,pady=9)
beaniesentry=Entry(childwareframe,font=("arial",15),bd=5,width=12)
beaniesentry.grid(row=1,column=5,pady=9,padx=10)
beaniesentry.insert(0,0)

mittenslabel=Label(childwareframe,text="Mittens",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
mittenslabel.grid(row=2,column=4,pady=9)
mittensentry=Entry(childwareframe,font=("arial",15),bd=5,width=12)
mittensentry.grid(row=2,column=5,pady=9,padx=10)
mittensentry.insert(0,0)

bootieslabel=Label(childwareframe,text="Booties",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
bootieslabel.grid(row=3,column=4,pady=9)
bootiesentry=Entry(childwareframe,font=("arial",15),bd=5,width=12)
bootiesentry.grid(row=3,column=5,pady=9,padx=10)
bootiesentry.insert(0,0)

sweaterlabel=Label(childwareframe,text="Sweaters",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
sweaterlabel.grid(row=4,column=4,pady=9)
sweaterentry=Entry(childwareframe,font=("arial",15),bd=5,width=12)
sweaterentry.grid(row=4,column=5,pady=9,padx=10)
sweaterentry.insert(0,0)

romperslabel=Label(childwareframe,text="Rompers",font=("Time New Roman",15,"bold"),bg="gray20",fg="white")
romperslabel.grid(row=5,column=4,pady=9)
rompersentry=Entry(childwareframe,font=("arial",15),bd=5,width=12)
rompersentry.grid(row=5,column=5,pady=9,padx=10)
rompersentry.insert(0,0)

# here is bill frame ////////////////////////////////////////////////////////////////////////////////////////////
billframe=Frame(productframe,bd=8,relief=GROOVE)
billframe.grid(row=0,column=5,padx=10)

billarealabel=Label(billframe,text="Bill Area",font=("Time New Roman",20,"bold"),bd=7,relief=GROOVE)
billarealabel.pack(fill=X)

# add scroll bar
scrollbar=Scrollbar(billframe,orient=VERTICAL)
scrollbar.pack(side=RIGHT,fill=Y)

textarea=Text(billframe,height=18,width=65,yscrollcommand=scrollbar.set)# connect with text area and scroll bar
textarea.pack()
scrollbar.config(command=textarea.yview)


# here is bill menu frame..................................................................................


billmenuframe=LabelFrame(root,text="BillMenu",font=("Time New Roman",15,"bold"),bg="gray20",fg="gold",bd=8,relief=GROOVE)
billmenuframe.pack(fill=X)

meanswareprizelabel=Label(billmenuframe,text="MeansWarePrize",font=("Time New Roman",14,"bold"),bg="gray20",fg="white")
meanswareprizelabel.grid(row=0,column=0,pady=9)
meanswareprizeentry=Entry(billmenuframe,font=("arial",14),bd=5,width=12)
meanswareprizeentry.grid(row=0,column=1,pady=9,padx=10)


womenswareprizelabel=Label(billmenuframe,text="WomensWarePrize",font=("Time New Roman",14,"bold"),bg="gray20",fg="white")
womenswareprizelabel.grid(row=1,column=0,pady=9)
womenswareprizeentry=Entry(billmenuframe,font=("arial",14),bd=5,width=12)
womenswareprizeentry.grid(row=1,column=1,pady=9,padx=10)


childwareprizelabel=Label(billmenuframe,text="ChildWarePrize",font=("Time New Roman",14,"bold"),bg="gray20",fg="white")
childwareprizelabel.grid(row=2,column=0,pady=9)
childwareprizeentry=Entry(billmenuframe,font=("arial",14),bd=5,width=12)
childwareprizeentry.grid(row=2,column=1,pady=9,padx=10)

# here is tax ...............................................................................................................


meanswareprizetaxlabel=Label(billmenuframe,text="MeanswarePrizetTax",font=("Time New Roman",14,"bold"),bg="gray20",fg="white")
meanswareprizetaxlabel.grid(row=0,column=2,pady=9)
meanswareprizetaxentry=Entry(billmenuframe,font=("arial",14),bd=5,width=12)
meanswareprizetaxentry.grid(row=0,column=3,pady=9,padx=10)

womenwareprizetaxlabel=Label(billmenuframe,text="WomenwarePrizetTax",font=("Time New Roman",14,"bold"),bg="gray20",fg="white")
womenwareprizetaxlabel.grid(row=1,column=2,pady=9)
womenwareprizetaxentry=Entry(billmenuframe,font=("arial",14),bd=5,width=12)
womenwareprizetaxentry.grid(row=1,column=3,pady=9,padx=10)

childwareprizetaxlabel=Label(billmenuframe,text="ChildwarePrizetTax",font=("Time New Roman",14,"bold"),bg="gray20",fg="white")
childwareprizetaxlabel.grid(row=2,column=2,pady=9)
childwareprizetaxentry=Entry(billmenuframe,font=("arial",14),bd=5,width=12)
childwareprizetaxentry.grid(row=2,column=3,pady=9,padx=10)

# here is button frame ..................................................................................................
buttonframe=Frame(billmenuframe,bd=8,relief=GROOVE)
buttonframe.grid(row=0,column=4,rowspan=3)

totalbutton=Button(buttonframe,text="Total",font=("arial",13,"bold"),bd=5,width=9,bg="gray20",fg="White",pady=10,command=total)
totalbutton.grid(row=0,column=4,pady=20,padx=6)

billbutton=Button(buttonframe,text="Bill",font=("arial",13,"bold"),bd=5,width=9,bg="gray20",fg="White",pady=10,command=bill)
billbutton.grid(row=0,column=5,pady=20,padx=5)

emailbutton=Button(buttonframe,text="Email",font=("arial",13,"bold"),bd=5,width=9,bg="gray20",fg="White",pady=10)
emailbutton.grid(row=0,column=6,pady=20,padx=5)

printbutton=Button(buttonframe,text="Print",font=("arial",13,"bold"),bd=5,width=9,bg="gray20",fg="White",pady=10,command=print_bill)
printbutton.grid(row=0,column=7,pady=20,padx=5)

clearbutton=Button(buttonframe,text="Clear",font=("arial",13,"bold"),bd=5,width=9,bg="gray20",fg="White",pady=10)
clearbutton.grid(row=0,column=8,pady=20,padx=5)



















root.mainloop()
```
